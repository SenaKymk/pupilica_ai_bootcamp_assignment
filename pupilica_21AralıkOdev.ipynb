```python
import pandas as pd

import matplotlib.pyplot as plt
veri = pd.read_csv("IRIS.csv")
veri
```

```python
plt.figure()
plt.hist(veri["sepal_length"], color = "red", alpha = 0.9, bins=20 , label = "sepal_length")
plt.hist(veri["petal_length"], color = "yellow", alpha = 0.9, bins=20 , label = "petal_length")

plt.ylabel("Frekans (KaÃ§ adet olduÄŸu)")
plt.xlabel("cm")
plt.legend()
```

```python
"""subplots"""
fig, ax = plt.subplots(2,1) # 2 satir, 1 sutun
fig.suptitle("Ana baslik")
ax[0].plot(veri.sepal_length, color = "red") # ax'lerin sifirincisi ilk eksen birinci plot
ax[0].set_title("Sepal Length")
ax[0].set_xlabel("index")
ax[0].set_ylabel("cm")

ax[1].plot(veri.petal_length, color = "blue")
ax[1].set_title("Petal Length")
ax[1].set_xlabel("index")
ax[1].set_ylabel("cm")

fig.tight_layout()
fig.subplots_adjust(top = 0.85)
plt.show()
```
