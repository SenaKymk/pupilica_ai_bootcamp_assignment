```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(11,5))
plt.plot(veri.index,veri["perimeter_mean"], color="blue",alpha=0.9 , label="perimeter_mean")

plt.plot(veri.index,veri["radius_mean"], color="#ff00ff",alpha=0.9 , label="radius_mean")
plt.title("perimeter_mean")
plt.xlabel("Veri no")
plt.ylabel("radius_mean")
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.scatter(veri.index,veri["perimeter_mean"], color="blue",alpha=0.9 ,s=15, label="perimeter_mean")
plt.scatter(veri.index,veri["radius_mean"], color="#ff00ff",alpha=0.9 ,s=15, label="radius_mean")
plt.title("perimeter_mean")
plt.xlabel("Veri no")
plt.ylabel("radius_mean")
plt.grid(True)
plt.legend()
plt.show()
"""pozitif bir korelasyon var diyemeyiz belli bi kısımda var belli bir kısımda yok

"""
```
